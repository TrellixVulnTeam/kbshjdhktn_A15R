```python
def dict_project_database_creation(data):
    """
    Функция принимает на вход базу данных с проектами, в которой есть колонки "Название проекта", "Описание проекта":
    Функиция возвращает словарь слов со второстепенным значением и добавляет в исходные данные колонки со словарями (тегами)
    и их общим количеством в документе
    """
    
    def dict_creation(array):
        cur_dict = {}
        for i_array in range(len(array)):
            for i_words in array[i_array]:
                if len(i_words)>2:
                    i_words = i_words[:-3]
                else:
                    i_words = ''
                cur_dict[i_words] = array[i_array].count(i_words)
        return cur_dict
    
    data['Описание проекта тэги'] = dict_creation(list(map(lambda x: str(x).strip().split(' '), data['Описание проекта'])))
    data['Название проекта тэги'] = dict_creation(list(map(lambda x: str(x).strip().split(' '), data['Название проекта'])))
    
    return data
```

```python
dict_project_database_creation(project)
```
